```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Score': [85, 90, 95, 80]}
df = pd.DataFrame(data)
df
```

```python
sorted_df = df.sort_values(by='Age')
print(sorted_df)
```

```python
sorted_df = df.sort_values(by='Age',ascending=False)
print(sorted_df)
```

```python
sorted_df = df.sort_values(by='Age',ascending=False, ignore_index=True)
print(sorted_df)
```

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Score': [85, 90, 95, 80]}
df = pd.DataFrame(data)
df
```

```python
# default는 ascending 순서
df_sorted_by_index = df.sort_index()
print(df_sorted_by_index)
```

```python
df_sorted_by_index_desc = df.sort_index(ascending=False)
print(df_sorted_by_index_desc)
```
