# Pandas DataFrame 기초

## 📚 용어 정리 (Terminology)

### 기본 데이터 구조 용어

| 용어 (한글) | 용어 (영문) | 설명 | 시각적 표현 |
|------------|-----------|------|-----------|
| **행** | Row / Index | 가로 방향의 데이터 줄. 위에서 아래로 날아감. | → (가로) |
| **열** | Column | 세로 방향의 데이터 줄. 왼쪽에서 오른쪽으로 진행. | ↓ (세로) |
| **셀** | Cell | 행과 열이 교차하는 하나의 데이터 단위 | 사각형 한 칸 |
| **인덱스** | Index | 각 행을 고유하게 식별하는 레이블 (0, 1, 2, ...) | 행의 '이름' |
| **컬럼명** | Column Name | 각 열을 식별하는 이름 ('Name', 'Age', 등) | 열의 '이름' |

### pandas 핵심 용어

| 용어 | 정의 | 비유 |
|------|------|------|
| **Series** | 1차원 배열 (한 줄의 데이터) | Excel의 한 열(Column) |
| **DataFrame** | 2차원 테이블 (행+열 구조) | Excel의 전체 시트 |
| **dtype** | 데이터 타입 (int, float, object 등) | 변수의 자료형 |
| **NaN** | Not a Number (결측값) | 빈칸, NULL |
| **레이블** | Label | 인덱스나 컬럼의 이름 | 주소, 이름표 |

### 방향 관련 용어

```
     컬럼명 (Column) → 가로 (Horizontal)
          ↓
    ┌─────┬─────┬─────┐
인덱스 │  A  │  B  │  C  │
(Row)├─────┼─────┼─────┤
  ↓  │  1  │  2  │  3  │  ← 데이터 셀 (Cell)
세로  ├─────┼─────┼─────┤
(Vertical)│  4  │  5  │  6  │
    └─────┴─────┴─────┘
```

### 중요 구분
- **Axis 0**: 행(Row) 방향, 위→아래, `axis=0` 또는 `axis='index'`
- **Axis 1**: 열(Column) 방향, 왼쪽→오른쪽, `axis=1` 또는 `axis='columns'`

## 3. 인덱싱과 슬라이싱

### 📚 인덱싱 용어 정리

| 용어 | 정의 | 예시 | 비고 |
|------|------|------|------|
| **인덱싱 (Indexing)** | 특정 데이터를 선택하여 가져오는 것 | `df[0]`, `df['Name']` | 위치나 이름으로 선택 |
| **슬라이싱 (Slicing)** | 범위를 지정하여 여러 데이터를 선택 | `df[0:3]`, `df['A':'C']` | 시작:끝 형태 |
| **레이블 (Label)** | 인덱스나 컬럼의 이름 | `'Name'`, `0`, `'Alice'` | 문자열 또는 숫자 |
| **위치 (Position)** | 0부터 시작하는 정수 순서 | 0, 1, 2, 3... | Python은 0부터 시작 |
| **마스크 (Mask)** | True/False로 구성된 불리언 배열 | `[True, False, True]` | 필터링에 사용 |

### 🎯 loc vs iloc 상세 비교

| 구분 | `loc` | `iloc` |
|------|-------|--------|
| **의미** | location (레이블 기반) | integer location (정수 위치 기반) |
| **사용 값** | 인덱스 레이블 (이름) | 0부터 시작하는 정수 위치 |
| **슬라이싱** | 끝값 **포함** `[a:b]` → a, b 모두 포함 | 끝값 **미포함** `[a:b]` → a만 포함 |
| **사용 예** | `df.loc['Alice']` | `df.iloc[0]` |
| **특징** | 이름으로 찾기 | 순서(번호)로 찾기 |

```
DataFrame 예시:
             Name    Age   (컬럼 레이블)
          ┌────────┬──────┐
0 (인덱스) │ 'Alice'│  30  │  ← 첫 번째 행 (위치 0)
          ├────────┼──────┤
1 (인덱스) │  'Bob' │  25  │  ← 두 번째 행 (위치 1)
          └────────┴──────┘

loc[0] → 인덱스 레이블이 0인 행 선택
iloc[0] → 위치 0번째(첫 번째) 행 선택
```

### 인덱스(Index)란?
- DataFrame의 각 행을 고유하게 식별하는 레이블
- 기본적으로 0, 1, 2, ... 와 같은 정수 인덱스가 자동 생성됨
- 사용자가 원하는 값으로 인덱스를 설정할 수 있음 (set_index)

### 주요 인덱싱 방법

| 방법 | 설명 | 예시 |
|------|------|------|
| `df['컬럼명']` | 단일 열 선택 | `df['Name']` |
| `df[['컬럼1', '컬럼2']]` | 여러 열 선택 | `df[['Name', 'Age']]` |
| `df.loc[]` | 레이블 기반 인덱싱 | `df.loc[0, 'Name']` |
| `df.iloc[]` | 위치 기반 인덱싱 (정수) | `df.iloc[0, 0]` |
| `df.set_index()` | 특정 열을 인덱스로 설정 | `df.set_index('Name')` |
| `df.reset_index()` | 인덱스를 기본값으로 리셋 | `df.reset_index()` |

### 📚 불리언 인덱싱 (Boolean Indexing) 용어 정리

| 용어 | 정의 | 예시 |
|------|------|------|
| **Boolean** | 참(True)/거짓(False) 두 값만 가지는 논리 타입 | `True`, `False` |
| **마스크 (Mask)** | 조건 결과에 따른 True/False 배열 | `[True, False, True]` |
| **필터링 (Filtering)** | 조건을 만족하는 데이터만 선택하는 것 | `df[df['Age'] > 20]` |
| **조건식** | 비교 연산을 포함한 식 | `df['Age'] > 25` |
| **논리 연산자** | 여러 조건을 결합하는 연산자 | `&` (AND), `|` (OR), `~` (NOT) |

### 비교 연산자 정리

| 연산자 | 의미 | 예시 | 결과 |
|--------|------|------|------|
| `>` | 크다 (greater than) | `5 > 3` | True |
| `<` | 작다 (less than) | `5 < 3` | False |
| `==` | 같다 (equal to) | `5 == 5` | True |
| `!=` | 같지 않다 (not equal) | `5 != 3` | True |
| `>=` | 크거나 같다 | `5 >= 5` | True |
| `<=` | 작거나 같다 | `3 <= 5` | True |

### 논리 연산자 정리 (주의: 괄호 필수!)

| 연산자 | 의미 | 설명 | 예시 |
|--------|------|------|------|
| `&` | AND (그리고) | 두 조건 모두 참이어야 True | `(df['Age'] > 20) & (df['Age'] < 30)` |
| `|` | OR (또는) | 둘 중 하나만 참이어도 True | `(df['Age'] < 20) | (df['Age'] > 60)` |
| `~` | NOT (부정) | True를 False로, False를 True로 | `~(df['Age'] > 20)` |

```python
# ⚠️ 중요: 각 조건은 반드시 괄호로 묶어야 함!
# 틀림: df['Age'] > 20 & df['Age'] < 30
# 맞음: (df['Age'] > 20) & (df['Age'] < 30)
```

### 불리언 인덱싱 과정

```
Step 1: 조건식 적용 → 마스크 생성
  df['Age'] > 25
  [25, 30, 22, 35] > 25
  → [False, True, False, True]  (마스크)

Step 2: 마스크를 DataFrame에 적용
  df[마스크]
  → True인 행만 선택됨
```

````python
# 불리언 인덱싱: Age가 25보다 큰 행들만 필터링
# 마스크를 DataFrame에 적용하여 True인 행만 선택

filtered_data = df[df['Age'] > 25]
print(filtered_data)
````
